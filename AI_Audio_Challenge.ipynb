```python
## importing the chat model of llm
GROQ_API_KEY=userdata.get("GROQ_!")
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.1-70b-versatile",
    temperature=0,
    max_tokens=30,
    timeout=None,
    max_retries=2,
    groq_api_key=GROQ_API_KEY

)
```

```python
## multiple models as option
print("Which model you wanna choose :")
print("1. LLaMA 3.1 70B Versatile")
print("2. llama-3.1-8b-instant")
print("3. mixtral-8x7b-32768")
print("4. whisper-large-v3")
choice = input("Enter your choice (1/2/3/4): ")
models = {
        "1": "llama-3.1-70b-versatile",
        "2": "llama-3.1-8b-instant",
        "3": "mixtral-8x7b-32768",
        "4": "whisper-large-v3"
    }
if choice in models:
        model_name = models[choice]
        model = ChatGroq(
            model=model_name,
            temperature=0,
            max_tokens=30,
            timeout=None,
            max_retries=2,
            groq_api_key=GROQ_API_KEY
        )
```

```python
## prompt
prompt =str(input("Enter your prompt : "))
result = model.invoke(prompt)
reply=result.content
```

```python
#importing the elevenlabs -- model = eleven_multilingual_v1

from elevenlabs import generate, play, set_api_key
from google.colab import userdata
client=set_api_key(
    api_key=userdata.get("ElevenLabs")
    )
audio = generate(
    text=reply,
    voice="Charlie",
    model="eleven_multilingual_v1"
)
```

```python
## end result
play(audio, notebook=True)
print(reply)
```
