## Lets try to add a vision model(llama 3.2 11b and 90b)

```python
# Function to encode the image
import base64
def encode_image(image_path):
  with open(image_path, "rb") as image_file:
    return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
# Path to your image
image_path = "/content/Screenshot (63).png" #enter the path
```

```python
#groq vision (llama 3.2 11b)
from groq import Groq

# Getting the base64 string
base64_image = encode_image(image_path) #encoding the image

client = Groq(
    api_key=userdata.get("GROQ_!")
)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": "track ships and count from this image"},  # can change the prompt from here
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{base64_image}",
                    },
                },
            ],
        }
    ],
    model="llama-3.2-11b-vision-preview",
)

image_description = (chat_completion.choices[0].message.content)
```

```python
# image to text output
print(image_description)
```
